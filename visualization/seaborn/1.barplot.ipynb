### seaborn
- 범주형 변수 사용 지원
- 복잡한 데이터 셋 구조를 간편하게 분석
- 내장된 테마로 matplotlib figure 스타일을 간단하게 제어
- 데이터에서 패턴을 나타내는 색상 팔레트를 선택하는 도구들
- 데이터 서브 셋을 비교하고 일변량이나 이변량 분포를 시각화하는 옵션

```python
#라이브러리
import seaborn as sns
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 한글처리
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False

# 경고 메세지
import warnings
warnings.simplefilter("ignore")
```

```python
# 폰트
sns.set(font_scale=1)
```

```python
# seaborn 스타일의 그래프를 사용
plt.style.use('seaborn')
```

```python
# 생성
movie_title = ['크루엘라','극장판 귀멸의 칼날:무한열차편','학교 가는 길']
audience = [664308,2099131,20067]
data={'영화제목':movie_title,"누적관객":audience}
df = pd.DataFrame(data)
df
```

### barplot

***matplotlib***

```python
# seaborn을 쓸 땐 한글 설정보다 먼저 실행
plt.bar(df['영화제목'], df['누적관객'])
```

```python
# 한글 설정 후 다시 실행
plt.bar(df['영화제목'], df['누적관객'])
```

```python
# 옵션
plt.bar(df['영화제목'], df['누적관객'], color=['blue','orange','green'])
plt.xlabel('영화제목')
plt.ylabel('누적관객')
```

***seaborn***

```python
# 알아서 범주 각각에 색상을 줌
# 라벨입력
sns.barplot(data=df,x='영화제목',y='누적관객')
```

- 누적관객 수를 기준으로 정렬

```python
sns.barplot(data=df,x='영화제목',y='누적관객', order=df.sort_values('누적관객').영화제목)
```

```python
sns.barplot(data=df,x='영화제목',y='누적관객', order=df.sort_values('누적관객', ascending=False).영화제목)
```

```python
chart = sns.barplot(data=df,x='영화제목',y='누적관객',
                    order=df.sort_values('누적관객', ascending=False).영화제목)

label_format = '{:,.0f}'

ylabels = [label_format.format(i) + '만 명' for i in chart.get_yticks() / 10000 ]
chart.set_yticklabels(ylabels)
```

***수평막대***

```python
# x, y 값 바꾸기
chart = sns.barplot(data=df,y='영화제목',x='누적관객',
                    order=df.sort_values('누적관객', ascending=False).영화제목)

label_format = '{:,.0f}'

xlabels = [label_format.format(i) + '만 명' for i in chart.get_xticks() / 10000 ]
chart.set_xticklabels(xlabels)
```

```python
# 세부 설정은 matplotlib
chart = sns.barplot(data=df,y='영화제목',x='누적관객',
                    order=df.sort_values('누적관객', ascending=False).영화제목)

label_format = '{:,.0f}'

xlabels = [label_format.format(i) + '만 명' for i in chart.get_xticks() / 10000 ]
chart.set_xticklabels(xlabels)

plt.xlabel('누적관객', fontsize=14)
plt.ylabel('영화제목', fontsize=14)
plt.title("영화 별 누적 관객 수", fontsize=18)
```
