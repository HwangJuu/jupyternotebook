```python
import seaborn as sns
sns.set(font_scale=1)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 한글처리
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False

# 경고 메세지
import warnings
warnings.simplefilter("ignore")
```

```python
df = sns.load_dataset('flights')
df.head()
```

### 데이터 탐색

```python
df.shape
```

```python
df.info()
```

```python
# 기본으로 오차범위를 같이 표시 해줌
sns.lineplot(data=df, x='year', y='passengers')
```

```python
sns.lineplot(data=df, x='year', y='passengers')
plt.xticks(df['year'], rotation=45);
```

####  1) hue : 그룹화

```python
# hue : 그룹으로 출력
sns.lineplot(data=df, x='year', y='passengers', hue='month')
```

#### 2) palette : 색상

```python
sns.lineplot(data=df, x='year', y='passengers', hue='month',palette='Set2')
```

#### 3) style : 선 모양

```python
sns.lineplot(data=df, x='year', y='passengers', hue='month',palette='Set2', style='month')
```

#### 4) marker

```python
sns.lineplot(data=df, x='year', y='passengers',
             hue='month',
             palette='Set2',
             style='month',
             markers=True)
```

#### 5) err_style : 오차 구간 스타일
- bars : 오차 구간 대신 오차 막대

```python
sns.lineplot(data=df, x='year', y='passengers', err_style='bars')
```
