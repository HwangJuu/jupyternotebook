### heatmap

- 색상으로 표현할 수 있는 다양한 정보를 일정한 이미지 위에 열 분포 형태로 출력

```python
import seaborn as sns
sns.set(font_scale=1)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 한글처리
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False

# 경고 메세지
import warnings
warnings.simplefilter("ignore")
```

```python
cm = [
    [2,0],
    [0,2]
]

sns.heatmap(cm)
```

```python
# cmap : 색상변경
sns.heatmap(cm, cmap='viridis')
```

```python
# annot : 데이터값 표시
sns.heatmap(cm, cmap='viridis', annot=True)
```

```python
df = sns.load_dataset('flights')
df.head()
```

#### 각 년도 별, 월별 승객 수 히트맵으로 시각화

- groupby() + 집계함수 => pivot_table => cross_table
- groupby, pivot

```python
pivot_data = df.pivot(index='month',columns='year', values='passengers')
pivot_data
```

```python
sns.heatmap(pivot_data)
```

```python
sns.heatmap(pivot_data, cmap='Blues')
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(pivot_data, cmap='Blues', annot=True)
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(pivot_data, cmap='Blues', annot=True, fmt='d')
```
