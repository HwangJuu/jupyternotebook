## pandas 차트 작성
- 참고 :  https://pandas.pydata.org/docs/user_guide/visualization.html

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# 한글처리
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False
```

```python
# 데이터 생성
df = pd.DataFrame(np.random.randn(1000,3), columns=list('ABC'))
# 누적합 : cumsum()
df = df.cumsum()

df.head()
```

#### 1) 기본 선 그래프 - 범례 기본

```python
# 기본 3가지 다 그려짐
df.plot();
```

```python
# B만 
df['B'].plot();
```

#### 2) 범례 제거

```python
df.plot(legend = False);
```

#### 3) 보조 축 설정

```python
df.plot(secondary_y = True);
```

```python
df['A'].plot()
df['B'].plot(secondary_y=True)
```

#### 4) 보조축 + 축 레이블 지정

```python
ax = df.plot(secondary_y=['A','B'])
ax.set_ylabel('C_scale')
ax.right_ax.set_ylabel('AB_Scale')
```

```python
ax = df.plot(secondary_y=['A','B'], mark_right = False)
ax.set_ylabel('C_scale')
ax.right_ax.set_ylabel('AB_Scale')
```

#### 5) subplots
- 나눠서 그려줌

```python
df.plot(subplots = True, figsize=(5,5))
```

```python
df.plot(subplots = True, figsize=(5,5), layout=(2,2))
```

```python
df.plot(subplots = True, figsize=(10,10), layout=(2,2), sharex = True)
```

```python
df.plot(subplots = True, figsize=(10,10), layout=(2,2), sharex = True, sharey = True)
```

```python
df.plot(subplots = True, figsize=(10,10), layout=(2,2), sharex = True, sharey = True)
plt.suptitle('Pandas Subplots')
```
