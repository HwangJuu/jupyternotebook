### pie plot(원 그래프)
- NaN이 있으면 0으로 채워짐
- 음수값이 있으면 오류 발생

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# 한글처리
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False
```

```python
ser = pd.Series(np.random.rand(4), index=list('abcd'), name='series')
ser
```

```python
ser.plot.pie()
```

```python
df = pd.DataFrame(np.random.rand(3,2), index=list('abc'), columns=['HA','HO'])
df
```

#### 컬럼이 여러개라서 선언을 해줘야함.
- pie requires either y column or 'subplots=True'

```python
df.plot.pie()
```

```python
# 하나의 값으로 그래프 
df.plot.pie(y='HA')
```

```python
df.plot.pie(subplots=True)
```

```python
# 사이즈 조절
df.plot.pie(subplots=True, figsize=(8,4))
```

```python
# 점유율
df.plot.pie(subplots=True, figsize=(8,4), autopct='%.2f%%')
```

```python
# 색상
df.plot.pie(subplots=True, figsize=(8,4), autopct='%.2f%%', colors=['r','g','b'])
```

```python
ser = pd.Series([0.1]*4, index=list('abcd'), name='series2')
ser
```

```python
# 1이 안되는 값을 그래프로 그리면 부채꼴 모양으로 나옴
ser.plot.pie(normalize=False)
```
