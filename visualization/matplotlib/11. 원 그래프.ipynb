```python
import matplotlib.pyplot as plt

plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False
```

- autopct : 원 안에 표시할 % 포맷
- startangle : 안 주는 경우 오른쪽 15도 정도에서 시작해서 시계 반대 방향으로 원이 돌아가게 됨, 각도를 주게 되면 그 각도를 기준으로 시계 반대 방향으로 원이 그려짐
- counterclock=False : 시계 방향으로 원이 작성되도록 함

#### [실습1]

```python
values = [30,25,20,13,10,2]
labels = ['Python','Java','JavaScript','C#','C/C++','etc']
```

```python
plt.pie(values, labels=labels)
```

```python
# 원 안에 표시 원이 차지하고 있는 비율 : autopct
plt.pie(values, labels=labels, autopct="%.1f%%")
plt.show()
```

```python
# 시작 각도 지정
plt.pie(values, labels=labels, autopct="%.1f%%", startangle=90)
plt.show()
```

```python
# 돌아가는 방향 조절
plt.pie(values, labels=labels, autopct="%.1f%%", startangle=90, counterclock=False)
plt.show()
```

#### [실습2] 원 조각 내기
- explode

```python
# 샘플
```

```python
# 각 조각에 대한 비율 지정(0은 조각 안냄)
# 조각 개수에 맞춰서 다 지정해 줘야함
explode = [0.2,0.1,0,0,0,0]

plt.pie(values, labels=labels, autopct="%.1f%%", startangle=90, counterclock=False, explode= explode)
plt.show()
```

## 심화

### 1) 색상 변경

```python
# 샘플
```

```python
# 갯수에 맞춰 지정
explode = [0.05]*6

colors=['#ffadad','#ffd6a5','#fdffb6','#caffbf','#9df6ff','#a0c4ff']

plt.pie(values, labels=labels, autopct="%.1f%%", startangle=90, counterclock=False,
        explode= explode, colors=colors)
plt.legend(loc=(1.2,0.3), title="언어별 선호도")
plt.show()
```

### 2) wedgeprops() : 도넛 모양 형태로 구성

```python
# 샘플
```

```python
explode = [0.05]*6

colors=['#ffadad','#ffd6a5','#fdffb6','#caffbf','#9df6ff','#a0c4ff']

# 가운데 모양 지정
wedgeprops = {'width':0.8}

plt.pie(values, labels=labels, autopct="%.1f%%", startangle=90, counterclock=False,
        explode= explode, colors=colors, wedgeprops=wedgeprops)
plt.legend(loc=(1.2,0.3), title="언어별 선호도")
plt.show()
```

#### explode 대신에 wedgeprops 이용

```python
colors=['#ffadad','#ffd6a5','#fdffb6','#caffbf','#9df6ff','#a0c4ff']

wedgeprops = {'width':0.8, 'edgecolor':'w','linewidth':3}

plt.pie(values, labels=labels, autopct="%.1f%%", startangle=90, counterclock=False,
        explode= explode, colors=colors, wedgeprops=wedgeprops)
plt.legend(loc=(1.2,0.3), title="언어별 선호도")
plt.show()
```

### 3) autopct 커스텀

```python
# 샘플
```

```python
# 함수만들기
# 파이썬 삼항연산자 
def custom_pct(pct):
    return ("%.1f%%" % pct) if pct >= 10 else ''
```

```python
colors=['#ffadad','#ffd6a5','#fdffb6','#caffbf','#9df6ff','#a0c4ff']

wedgeprops = {'width':0.8, 'edgecolor':'w','linewidth':3}

plt.pie(values, labels=labels, autopct=custom_pct, startangle=90, counterclock=False,
        explode= explode, colors=colors, wedgeprops=wedgeprops)
plt.legend(loc=(1.2,0.3), title="언어별 선호도")
plt.show()
```
