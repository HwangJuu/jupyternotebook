```python
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False
```

```python
# 생성
data = {
    "이름" : ['채치수','정대만','송태섭','서태웅','강백호','변덕규','황태산','윤대협'],
    "학교" : ['북산고','북산고','북산고','북산고','북산고','능남고','능남고','능남고'],
    "키" :[197,184,168,187,188,202,188,190],
    "국어" : [90,40,80,40,15,80,55,100],
    "영어" : [85,35,75,60,20,100,65,85],
    "수학" : [100,50,70,70,10,95,45,90],
    "과학" : [95,55,80,75,35,85,40,95],
    "사회" : [85,25,75,80,10,80,35,95],
    "SW특기" : ['Python', 'Java','Javascript', '', '', 'C', 'PYTHON', 'C#']
}
student_df = pd.DataFrame(data)
student_df
```

```python
# x축 : 이름, y축 : 국어 점수 --> 선 그래프
plt.plot(student_df['이름'], student_df['국어'])
```

```python
plt.plot('이름','국어', data=student_df)
```

```python
# 영어 점수로 bar chart
plt.bar('이름','영어', data=student_df)
```

```python
# 영어 점수로 bar chart
plt.barh('이름','영어', data=student_df)
```

```python
# 영어점수 내림차순
student_df.sort_values('영어')
```

```python
# 영어 점수로 bar chart
plt.barh('이름','영어', data=student_df.sort_values('영어'))
```

```python
# 국,영,수 성적
plt.plot('이름','국어',data=student_df, label="국어")
plt.plot('이름','영어',data=student_df, label="영어")
plt.plot('이름','수학',data=student_df, label="수학")
plt.legend(ncol=3)
plt.show()
```

```python
# 사이즈 업
plt.figure(figsize=(10,5))

plt.plot('이름','국어',data=student_df, label="국어")
plt.plot('이름','영어',data=student_df, label="영어")
plt.plot('이름','수학',data=student_df, label="수학")
plt.legend(ncol=3)
plt.show()
```

```python
# 그리드 추가
plt.figure(figsize=(10,5))

plt.plot('이름','국어',data=student_df, label="국어")
plt.plot('이름','영어',data=student_df, label="영어")
plt.plot('이름','수학',data=student_df, label="수학")
plt.legend(ncol=3)
plt.grid()

plt.show()
```

```python
# y축만 그리드 추가
plt.figure(figsize=(10,5))

plt.plot('이름','국어',data=student_df, label="국어")
plt.plot('이름','영어',data=student_df, label="영어")
plt.plot('이름','수학',data=student_df, label="수학")
plt.legend(ncol=3)
# 그리드 효과 주기 가능
plt.grid(axis = 'y', alpha=0.3, color='red', linestyle='--', linewidth=2)

plt.show()
```
