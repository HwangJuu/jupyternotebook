```python
print(" : %f " % starwars['height'].skew())
print(" : %f " % starwars['height'].kurt())
```

```python
plt.style.use('seaborn')
fig, ax = plt.subplots(1,2, figsize=(16,5))

sns.histplot(starwars[starwars['sex'] == 'male']['height'], ax=ax[0],
                                                         kde=True, stat='density', color='lightpink')
sns.histplot(starwars[starwars['sex'] == 'female']['height'], ax=ax[0],
                                                         kde=True, stat='density', color='blue')
sns.histplot(starwars[starwars['sex'] == 'robot']['height'], ax=ax[0],
                                                         kde=True, stat='density')
ax[0].legend(['male','female','robot'])

# kdeplot :   
sns.kdeplot(starwars[starwars['gender'] == 'masculine']['height'], ax=ax[1])
sns.kdeplot(starwars[starwars['gender'] == 'feminine']['height'], ax=ax[1])
ax[1].legend(['masculine','feminine'])
```
