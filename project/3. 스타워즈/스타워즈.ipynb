#### 2) 스타워즈 캐릭터의 성별에 따른 신장의 분포 알아보기

```python
starwars['height'].describe()
```

```python
plt.style.use('seaborn')
fig, ax = plt.subplots(1,2, figsize=(16,5))

sns.histplot(starwars[starwars['sex'] == 'male']['height'], ax=ax[0],
                                                         kde=True, stat='density', color='lightpink')
sns.histplot(starwars[starwars['sex'] == 'female']['height'], ax=ax[0],
                                                         kde=True, stat='density', color='blue')
sns.histplot(starwars[starwars['sex'] == 'robot']['height'], ax=ax[0],
                                                         kde=True, stat='density')
ax[0].legend(['male','female','robot'])

# kdeplot : 확률분포를 부드럽게 그려줌
sns.kdeplot(starwars[starwars['gender'] == 'masculine']['height'], ax=ax[1])
sns.kdeplot(starwars[starwars['gender'] == 'feminine']['height'], ax=ax[1])
ax[1].legend(['masculine','feminine'])
```

#### 4) 스타워즈 캐릭터의 종족간 몸무게 비교

```python
# 그룹 잡기 
weight = starwars[['species','mass']].groupby('species').mean()
weight.head(3)
```

***그래프***

- pandas

```python
# 몸무게 순서
weight['mass'].sort_values(ascending=False)
```

```python
# 상위 10개
weight['mass'].sort_values(ascending=False).head(10)
```

```python
# 그래프 그리기 - 색상 옵션 및 이름 기울기
weight['mass'].sort_values(ascending=False).head(10).plot.bar(color='lightgreen', rot =45)
```

#### 5) 스타워즈 캐릭터의 몸무게 - 키 상관관계

- 제외
    - 몸무게 == 1358, 0

```python
starwars.shape
```

```python
# 몸무게 1358 
out1 = ~starwars['species'].isin(['Hutt'])
```

```python
# 몸무게 0
out2 = ~ starwars['mass'].isin([0.0])
```

```python
starwars.loc[out1 & out2].shape
```

```python
sns.regplot(data=starwars.loc[out1 & out2], x='height', y='mass')
```

```python
sns.lmplot(data=starwars.loc[out1 & out2], x='height', y='mass')
```

```python
sns.lmplot(data=starwars.loc[out1 & out2], x='height', y='mass', hue='gender')
```
