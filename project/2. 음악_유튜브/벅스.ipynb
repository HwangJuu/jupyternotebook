#### 1) 셀레니움 드라이버 로드

```python
from selenium import webdriver
from bs4 import BeautifulSoup
import time
```

```python
url = 'https://music.bugs.co.kr/chart'
```

```python
browser = webdriver.Chrome('../../chromedriver')
browser.get(url)

time.sleep(3)

soup = BeautifulSoup(browser.page_source,"lxml")
```

#### 벅스 음악 정보 가져오기
- 순위, 노래제목, 가수이름

```python
# 데이터 확인
songs = soup.select("table.byChart > tbody > tr")
print(len(songs))
print(songs[0])
```

```python
for idx, song in enumerate(songs):
    title = song.select_one('p.title > a').text
    singer = song.select_one('p.artist > a').text
    
    print(idx+1, title, singer)
```

#### 크롤링 결과 엑셀 저장

```python
bugs_list = []

for idx, song in enumerate(songs):
    title = song.select_one('p.title > a').text
    singer = song.select_one('p.artist > a').text
    
    print(idx+1, title, singer)
    
    bugs_list.append(['벅스',idx+1, title, singer])
```

```python
import pandas as pd
```

```python
colums = ['서비스','순위','타이틀','가수']

df = pd.DataFrame(bugs_list, columns= colums)
df.head()
```

```python
df.to_excel("../../data/dataset/bugs.xlsx", index=False)
```
