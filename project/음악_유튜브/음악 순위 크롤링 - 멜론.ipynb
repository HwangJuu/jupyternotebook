### 멜론, 벅스, 지니 음악 순위 크롤링

- 각 음악 사이트의 크롤링 정보를 엑셀로 저장
- 저장된 정보를 읽어온 후 3개의 데이터 합치기

#### 1) 셀레니움 드라이버 로드

```python
from selenium import webdriver
from bs4 import BeautifulSoup
import time
```

```python
url = 'https://www.melon.com/chart/index.htm'
```

```python
browser = webdriver.Chrome('../../chromedriver')
browser.get(url)

time.sleep(3)

soup = BeautifulSoup(browser.page_source,"lxml")
```

### 멜론 음악 정보 가져오기
- 순위, 노래제목, 가수이름 

```python
# 전체 데이터 확인
songs = soup.select("table > tbody > tr")
print(len(songs))
print(songs[0])
```

```python
for idx,song in enumerate(songs):
    # 제목 : 클래스명 이용
    title = song.select_one("div.ellipsis.rank01 > span > a").text
    # 가수 
    singer = song.select_one("div.ellipsis.rank02 > a").text
    
    print(idx+1, title,singer)
```

#### 크롤링 결과 엑셀 저장
- 서비스 : 멜론
- 순위, 타이틀, 가수 순으로 저장

```python
# 리스트로 담기
melon_list = []
for idx,song in enumerate(songs):
    # 제목 : 클래스명 이용
    title = song.select_one("div.ellipsis.rank01 > span > a").text
    # 가수 
    singer = song.select_one("div.ellipsis.rank02 > a").text
    
    print(idx+1, title, singer)
    melon_list.append(["멜론",idx+1, title, singer])
```

```python
import pandas as pd
```

```python
colums = ['서비스','순위','타이틀','가수']

df = pd.DataFrame(melon_list, columns= colums)
df.head()
```

```python
df.to_excel("../../data/dataset/melon.xlsx", index=False)
```
